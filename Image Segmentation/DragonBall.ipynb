```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.cluster import KMeans
import cv2
```

```python
dragon=cv2.imread('dragon_ball.png')
dragon=cv2.cvtColor(dragon,cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(dragon)
```

```python
dragon=cv2.resize(dragon,(dragon.shape[1]//4,dragon.shape[0]//4))
```

```python
dragon.shape
```

```python
model=KMeans(20)
```

```python
X=dragon.reshape(-1,3)
```

```python
model.fit(X)
```

```python
pallets=model.cluster_centers_/255
```

```python
pallets=pallets.reshape(1,-1,3)
plt.imshow(pallets)
```

```python
duplicate=np.zeros(dragon.shape)
```

```python
dragon.shape[:2]
```

```python
vals = model.labels_.reshape(dragon.shape[:2])
```

```python
for row in range(vals.shape[0]):
    for col in range(vals.shape[1]):
        duplicate[row, col] = pallets[0][vals[row, col]]
```
